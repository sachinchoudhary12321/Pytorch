```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)

w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
def loss(prediction, target):
    """Binary cross entropy loss function with clipping for stability."""
    epsilon = 1e-8
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -target * torch.log(prediction) - (1 - target) * torch.log(1 - prediction)
```

```python
z = w * x + b
y_pred = torch.sigmoid(z)

# Changed variable name from 'loss' to 'first_loss_val' to prevent overwriting the function
first_loss_val = loss(y_pred, y)
print(first_loss_val)
```

```python
loss
```

```python
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))

dy_pred_dz = y_pred*(1-y_pred)

dz_dw = x

dz_db = 1

dl_dw = dloss_dy_pred*dy_pred_dz*dz_dw

dl_db = dloss_dy_pred*dy_pred_dz*dz_db
```

```python
print(dl_dw)
print(dl_db)
```
