```python
class Model(nn.Module):
  def __init__(self, nums_feature):
    super().__init__()
    # it is the first layer which is take the 5 input and output is 3
    self.linear1 = nn.Linear(nums_feature, 3)

    self.relu = nn.ReLU()
# hidden layer ip=3,op=1
    self.linear2 = nn.Linear(3, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, feature):
    out = self.linear1(feature)
    out = self.relu(out)
    out = self.linear2(out)
    out = self.sigmoid(out)
    return out
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace = True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor = torch.tensor(X_train,dtype=torch.float32)
X_test_tensor = torch.tensor(X_test,dtype=torch.float32)
y_train_tensor = torch.tensor(y_train,dtype=torch.float32)
y_test_tensor = torch.tensor(y_test,dtype=torch.float32)
```

```python
class MYSimpleNN(nn.Module):
  def __init__(self, nums_feature):
    super().__init__()

    self.network = nn.Sequential(
    nn.Linear(nums_feature, 3),
    nn.ReLU(),
# hidden layer ip=3,op=1
    nn.Linear(3, 1),
    nn.Sigmoid(),
    )
    # it is the first layer which is take the 5 input and output is 3


  def forward(self, feature):
    out = self.network(feature)
    return out

  def loss(self,y_pred,y_true):
    return torch.mean((y_pred-y_true)**2)
```

```python
learning_rate = 0.1
epochs = 20
model = MYSimpleNN(X_train_tensor.shape[1])
model = model.to(torch.float64) # Add this line to cast the model to float64

for epoch in range(epochs):
  model.zero_grad() # Zero the gradients before each epoch
  #forward
  # Cast X_train_tensor to float64 to match model.weights and model.bias dtype
  y_pred = model(X_train_tensor.to(torch.float64))
  loss = model.loss(y_pred,y_train_tensor.to(torch.float64))
  print(f'epoch{epoch}:loss{loss}')
  #backward
  loss.backward()

  with torch.no_grad():
    for param in model.parameters():
      if param.grad is not None:
        param -= learning_rate * param.grad


with torch.no_grad():
  y_pred = model.forward(X_test_tensor.to(torch.float64))
  y_pred = (y_pred>0.5).float()
  y_pred = y_pred.numpy()
```

```python
learning_rate = 0.1
epochs = 20
model = MYSimpleNN(X_train_tensor.shape[1])
model = model.to(torch.float64) # Add this line to cast the model to float64

optimizer = torch.optim.SGD(model.parameters(),lr = learning_rate)

for epoch in range(epochs):
  model.zero_grad() # Zero the gradients before each epoch
  #forward
  # Cast X_train_tensor to float64 to match model.weights and model.bias dtype
  y_pred = model(X_train_tensor.to(torch.float64))
  loss = loss_fn(y_pred,y_train_tensor.to(torch.float64).unsqueeze(1)) # Reshape y_train_tensor
  print(f'epoch{epoch}:loss{loss}')
  #backward
  optimizer.zero_grad()

  loss.backward()

  optimizer.step()

```

# mini batch gradient descent

```python
batch_size = 100
epochs = 20
n_samples = len(X_train_tensor)

for epoch in range(epochs):

  for start_idx in range(0,n_samples,batch_size):

    end_idx = start_idx + batch_size;
    X_batch = X_train_tensor[start_idx:end_idx]
    y_batch = y_train_tensor[start_idx:end_idx]

    # Zero the gradients before each epoch
  #forward
  # Cast X_train_tensor to float64 to match model.weights and model.bias dtype
    y_pred = model(X_batch.to(torch.float64))
    loss = loss_fn(y_pred,y_batch.to(torch.float64).unsqueeze(1)) # Reshape y_train_tensor
    print(f'epoch{epoch}:loss{loss}')
    #backward
    optimizer.zero_grad()

    loss.backward()

    optimizer.step()

```

This code snippet demonstrates a training loop for a neural network using mini-batch gradient descent. Let's break it down:

### 1. Initialization

```python
batch_size = 100
epochs = 20
n_samples = len(X_train_tensor)
```

- **`batch_size = 100`**: Defines how many samples from the training data will be processed at once in each iteration of the inner loop. Using batches helps in faster training and more stable gradient updates.
- **`epochs = 20`**: Specifies the number of times the entire training dataset will be passed forward and backward through the neural network. One epoch means the model has seen the entire training dataset once.
- **`n_samples = len(X_train_tensor)`**: Gets the total number of samples in your training dataset (`X_train_tensor`). This is used to determine the range for iterating through batches.

### 2. Training Loop Structure

```python
for epoch in range(epochs):

  for start_idx in range(0,n_samples,batch_size):

    end_idx = start_idx + batch_size;
    X_batch = X_train_tensor[start_idx:end_idx]
    y_batch = y_train_tensor[start_idx:end_idx]
```

- **`for epoch in range(epochs):`**: This is the outer loop, which iterates `epochs` number of times. In each epoch, the model processes the entire training dataset.
- **`for start_idx in range(0,n_samples,batch_size):`**: This is the inner loop. It iterates through the training data in steps of `batch_size`.
  - `start_idx` marks the beginning index of the current batch.
  - `end_idx = start_idx + batch_size` marks the end index of the current batch.
  - `X_batch` and `y_batch` are created by slicing `X_train_tensor` and `y_train_tensor` to get the current mini-batch of features and labels, respectively.

### 3. Forward Pass and Loss Calculation

```python
    y_pred = model(X_batch.to(torch.float64))
    loss = loss_fn(y_pred,y_batch.to(torch.float64).unsqueeze(1)) # Reshape y_train_tensor
    print(f'epoch{epoch}:loss{loss}')
```

- **`y_pred = model(X_batch.to(torch.float64))`**: This performs the **forward pass**. The `X_batch` (input features for the current batch) is passed through the `model`. `.to(torch.float64)` ensures that the input data type matches the model's expected data type (which was cast to `float64` earlier in the notebook).
- **`loss = loss_fn(y_pred,y_batch.to(torch.float64).unsqueeze(1))`**: This calculates the **loss**. The `loss_fn` (Binary Cross-Entropy Loss, `nn.BCELoss()`, as defined in a previous cell) compares the model's predictions (`y_pred`) with the actual labels (`y_batch`).
  - `.unsqueeze(1)` is used to reshape `y_batch` to have a dimension of `[batch_size, 1]`, matching the output shape of `y_pred` from the model's final `nn.Linear` layer.
- **`print(f'epoch{epoch}:loss{loss}')`**: Prints the current epoch number and the calculated loss for the current batch.

### 4. Backward Pass and Optimization

```python
    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```

- **`optimizer.zero_grad()`**: Before calculating gradients for the current batch, it's crucial to zero out any previously accumulated gradients. This prevents gradients from previous batches or iterations from interfering with the current update.
- **`loss.backward()`**: This performs the **backward pass** (backpropagation). It computes the gradients of the `loss` with respect to all trainable parameters in the model. These gradients indicate how much each parameter should be adjusted to reduce the loss.
- **`optimizer.step()`**: This updates the model's parameters using the calculated gradients. The `optimizer` (Stochastic Gradient Descent, `torch.optim.SGD`, as defined in a previous cell) applies the updates based on its configured learning rate to move the parameters in the direction that minimizes the loss.

In summary, this code iteratively feeds mini-batches of data to the neural network, calculates the prediction error, and then adjusts the network's weights and biases to minimize that error over multiple epochs.

```python
batch_size = 100
epochs = 20
n_samples = len(X_train_tensor)

for epoch in range(epochs):

  for start_idx in range(0,n_samples,batch_size):

    end_idx = start_idx + batch_size;
    X_batch = X_train_tensor[start_idx:end_idx]
    y_batch = y_train_tensor[start_idx:end_idx]

    # Zero the gradients before each epoch
  #forward
  # Cast X_train_tensor to float64 to match model.weights and model.bias dtype
    y_pred = model(X_batch.to(torch.float64))
    loss = loss_fn(y_pred,y_batch.to(torch.float64).unsqueeze(1)) # Reshape y_train_tensor
    print(f'epoch{epoch}:loss{loss}')
    #backward
    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```
