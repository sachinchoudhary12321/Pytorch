#pipeline using dataset and dataloader

```python
from torch.utils.data import Dataset,DataLoader
```

```python
class CustomDataset(Dataset):# here we fill the functions
    def __init__(self,features,labels):
        self.feature = features
        self.label = labels

    def __len__(self):
        return len(self.feature)

    def __getitem__(self,idx):
        return self.feature[idx],self.label[idx]
```

```python
train_dataset = CustomDataset(X_train_tensor,y_train_tensor)
test_dataset = CustomDataset(X_test_tensor,y_train_tensor)
```

# define the model

```python
import torch.nn as nn

class MYSimpleNN(nn.Module):
  def __init__(self,X_input_features):
    super().__init__()
    self.weights = nn.Parameter(torch.rand(X_input_features,1,dtype=torch.float64))
    self.bias = nn.Parameter(torch.zeros(1,dtype=torch.float64))

  def forward(self,X):
    return torch.sigmoid(torch.matmul(X,self.weights)+self.bias)
```

```python
learning_rate = 0.1
epochs = 20
```

```python
#create model
model = MYSimpleNN(X_train_tensor.shape[1])

optimizer = torch.optim.SGD(model.parameters(),lr=learning_rate)

loss_function = nn.BCELoss()
```

```python
model.bias
```

```python
import torch

# forward pass
# loss calculate
# backward pass
#parametr update

for epoch in range(epochs):
  #forward
  # Cast X_train_tensor to float64 to match model.weights and model.bias dtype
  y_pred = model.forward(X_train_tensor.to(torch.float64))



  loss = model.loss(y_pred,y_train_tensor.to(torch.float64))



  #backward
  loss.backward()

  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

    model.weights.grad.zero_()
    model.bias.grad.zero_()

    print(f'epoch:{epoch+1},loss:{loss.item()}')


  #update
```

```python
with torch.no_grad():
  y_pred = model.forward(X_test_tensor.to(torch.float64))
  y_pred = (y_pred>0.5).float()
  y_pred = y_pred.numpy()
  print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```
